```python
# importing all neccessary librarys

import cv2
import os
import numpy as np
import pickle
```

```python
# mapping all names into numbers for processing

def name_map(x):
    if x=='brandon curry':
        return 0
    if x=='Dexter Jackson':
        return 1
    if x=='Hadi':
        return 2
    if x=='Jay Cutler':
        return 3
    if x=='Larry Scott':
        return 4
    if x=='Mamdouh Elssbiay':
        return 5
    if x=='Phil Heath':
        return 6
    if x=='Ronnie Coleman':
        return 7
    if x=='Shawn Rhoden':
        return 8
```

```python
# Mapping numbers into name for getting name after prediction

def get_name(x):
    if x==0:
        return 'brandon curry'
    if x==1:
        return 'Dexter Jackson'
    if x==2:
        return 'Hadi'
    if x==3:
        return 'Jay Cutler'
    if x==4:
        return 'Larry Scott'
    if x==5:
        return 'Mamdouh Elssbiay'
    if x==6:
        return 'Phil Heath'
    if x==7:
        return 'Ronnie Coleman'
    if x==8:
        return 'Shawn Rhoden'
```

```python
# Function for Detecting face 


def detect_face(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    
    face_cascade = cv2.CascadeClassifier('/home/user/Desktop/face_recognition/haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)
    
    if (len(faces) == 0):
        return None, None
    

    (x, y, w, h) = faces[0]
    
    return gray[y:y+w, x:x+h], faces[0]
```

```python
# Training the data 

def prepare_training_data(data_folder_path):

    dirs=os.listdir(data_folder_path)

    faces=[] # Detected faces stored as a numerical list
    labels=[] # name of the directorys that we stored images


    for dir_name in dirs:
        
        label=dir_name


        subject_dir_path=data_folder_path + '/'+ dir_name
        subject_img_names =os.listdir(subject_dir_path)
        
        for img_name in subject_img_names:
            img_path=subject_dir_path + '/'+img_name

            image=cv2.imread(img_path)

            cv2.imshow('training on image..',image)
            cv2.waitKey(1000)

            #detect face 

            face,rect=detect_face(image)

            if face is not None:

                faces.append(face)
                labels.append(name_map(label))

                print(labels,faces)

    cv2.destroyAllWindows()
    cv2.waitKey(1)                

    return faces,labels
```

```python

print('preparing data..')
faces,labels=prepare_training_data('/home/user/Desktop/face_recognition/DB')

print('data prepared.')
```

```python
print('total face:',len(faces))
```

```python
print('total labels :',len(labels))
```

```python
face_recognizer=cv2.face.LBPHFaceRecognizer_create()
```

```python
face_recognizer.train(faces,np.array(labels))
```

```python
def draw_rectangle(img,rect):
    (x,y,w,h)=rect
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
def draw_text(img,text,x,y):
    cv2.putText(img,text,(x,y),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,225),2)
```

```python
def predict(test_img):
    img=test_img.copy()


    face,rect=detect_face(img)

    label=face_recognizer.predict(face)
    label_text=get_name(label[0])
    s = label_text
    print(s)
    draw_rectangle(img,rect)

    draw_text(img,label_text,rect[0],rect[1])

    return img    
```
